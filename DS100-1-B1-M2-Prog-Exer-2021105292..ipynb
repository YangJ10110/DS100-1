## BAR GRAPH: Frequency vs Number of Allergens

```python
allergen_counts = data["Allergens"].str.split(",").apply(len); #counting the Allergen frequency
plt.figure(figsize=(10,6)) # setting up the figure size of the graph 10x6
sns.barplot(x=allergen_counts.value_counts().index, y=allergen_counts.value_counts().values) #setting up the values of the y and x axis
plt.xlabel("Number of Allergens")
plt.ylabel("Frequency")
plt.title("Frequency of Allergens per food item")
plt.show()
```

    The bar graph visually depicts the distribution of allergens per food item in the dataset, showcasing the frequency of occurrence for different numbers of allergens. The x-axis represents the number of allergens, while the y-axis quantifies the count or frequency of food items associated with each specific number of allergens.

    Each bar in the graph corresponds to a distinct number of allergens, and its vertical extent denotes the frequency of food items exhibiting that specific allergen count. Analysis of the graph reveals significant insights into the prevalence and distribution patterns of allergens in the dataset. It indicates that a substantial proportion of food items tend to contain a single allergen, whereas the occurrence of multiple allergens within a single dish is comparatively less frequent.

    This observation offers valuable implications for understanding the composition and complexity of allergens in various food products. It aids in assessing the risks associated with allergenic ingredients and supports the development of effective strategies to manage food allergies. The graphical representation of these findings provides a concise and intuitive summary of the dataset, facilitating quick comprehension and analysis of allergen frequencies.

# PIE CHART: ALLERGEN DISTRIBUTION

```python

allergens = data["Allergens"].str.split(", ").explode().dropna()
allergens = allergens[allergens != "None"] #filter out the none in the allergens
allergen_counts = allergens.value_counts()

plt.pie(allergen_counts.values, labels=allergen_counts.index, autopct="%1.1f%%")
plt.title("Allergen Distribution")
plt.axis("equal")
plt.show()
```

    The pie chart graph illustrates the distribution of allergens in the given dataset. Each slice of the pie represents a specific allergen, and its size corresponds to the proportion or percentage of occurrence in the dataset.

    By observing the trend of the pie chart we can See the relative frequencies of different allergens and their contribution to the overall allergen distribution. The larger the slice such as Dairy and Wheat, the more prevalent the corresponding allergen is in the dataset.

    The significance of this visualization is to provide a visual representation of the relative prevalence of different allergens in the food ingredients. It allows us to identify the allergens that are most commonly associated with the dataset and gain insights into the composition of allergens in food products. This information is valuable for individuals with food allergies, researchers, and food manufacturers, as it helps in understanding and managing potential allergenic risks.

# LINE GRAPH: FREQUENCY VS MAIN ALLERGENS

```python
data = pd.read_csv("food_ingredients_and_allergens.csv")

allergens_counts = data["Allergens"].value_counts()
allergens_counts = allergens_counts[allergens_counts.index != "None"]  # Filter out the "None" value

plt.figure(figsize=(20, 6))  # Set the desired width and height of the figure
allergens_counts.sort_index().plot(kind="line")

plt.xlabel("Allergens")
plt.ylabel("Frequency")
plt.title("Frequency of Allergens")
plt.show()
```

## LINE GRAPH INTERCTIVE: FREQUENCY VS ALLERGENS

```python
data = pd.read_csv("food_ingredients_and_allergens.csv")

allergens_counts = data["Allergens"].value_counts().sort_index()
allergens_counts = allergens_counts[allergens_counts.index != "None"]  # Filter out the "None" value

allergens_counts = allergens_counts.reset_index()  # Reset the index

fig = px.line(allergens_counts, x="index", y=allergens_counts["Allergens"],
              labels={"index": "Allergens", "y": "Frequency"},
              title="Frequency of Allergens")

fig.update_traces(hovertemplate="Allergen: %{x}<br>Frequency: %{y}")

fig.show()
```

    The line graph depicts the frequency of allergens in the given dataset. The x-axis represents different allergens, while the y-axis represents the frequency of occurrence. The graph shows how often each allergen appears in the dataset.

    It can be observed that the relative frequencies of different allergens. Each point on the line represents a specific allergen, and its vertical position indicates the frequency at which it appears in the dataset. By examining the slope and peaks of the line, we can identify the allergens that occur most frequently.
    
    The significance of this visualization is to provide insights into the prevalence of different allergens in the food ingredients. It helps identify the allergens that are more commonly associated with food products, enabling researchers, food manufacturers, healthcare professionals, and individuals with food allergies to better understand and manage potential allergenic risks.
